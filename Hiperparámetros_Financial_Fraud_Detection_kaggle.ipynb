## 2.3 Creación de Nuevas variables: Feature Engineering

## 2.3.1 <ins> Variable **'type'**:</ins>

* Creamos una nueva variable, llamada **"type_2"**, uniendo las primeras letras de las dos columnas, **"nameOrig"** y **"nameDest"**, los cuales nos representan el cliente que inicio la operación y el cliente que recibió la transacción

```python
## Utilizamos expresiones regulares para lograr este resultado - También colocamos otra forma de hacerlo

# df_banco['type_2'] = df_banco['nameOrig'].apply(lambda x:x[0]) + df_banco['nameDest'].apply(lambda x:x[0])
# df_banco['type_2'] = df_banco['nameOrig'].apply(lambda x:re.search('[A-Za-z]', x).group()) + df_banco['nameDest'].apply(lambda x:re.search('[A-Za-z]', x).group())
# df_banco['type_2']

df_banco = df_banco.with_columns((pl.col("nameOrig").apply(lambda x:x[0]) + pl.col("nameDest").apply(lambda x:x[0])).alias("type_2"))
```

```python
## Validamos que se haya creado correctamente la nueva variable
df_banco.head()
```

## 2.3.2 <ins> Variable **'Step'**:</ins>

* Tenemos como información que esta variable **'step'**, nos representa una unidad de tiempo, donde 1 step equivale a 1 hora y asi sucesivamente.
* Por consiguiente, agruparemos esta variable cada 24 horas para crear una nueva variable que nos represente los días.

```python
## Verificando todos los valores únicos de la variable 'step'
df_banco.select('step').unique().to_pandas().shape[0]
```

```python
## Creamos una nueva columna 'day', la cual agrupará cada 24 horas los steps
df_banco = df_banco.with_columns((pl.col('step').apply(lambda x: (x//24 +1))).alias('day'))
df_banco.select('day').to_pandas()
```

```python
## Validamos que se haya creado correctamente la nueva variable
df_banco.head()
```

```python
## Función para asignar la parte del día
def part_of_the_day(hora):
    hora_del_dia = hora % 24
    if 6 <= hora_del_dia < 12:
        return 'mañana'
    elif 12 <= hora_del_dia < 18:
        return 'tarde'
    elif 18 <= hora_del_dia < 24:
        return 'noche'
    else:
        return 'madrugada'

## Creamos la nueva columna con las partes del día correspondientes: 'mañana' / 'tarde' / 'noche' / 'madrugada'
df_banco = df_banco.with_columns((pl.col('step').apply(part_of_the_day)).alias('part_of_the_day'))
df_banco.select('part_of_the_day').to_pandas()
```

```python
## Validamos que se haya creado correctamente la nueva variable
df_banco
```

## 2.4 Eliminando Variables no relevantes


* Luego de analizar las variables y el origen de nuestro dataset, existe una nota importante a tener en cuenta, donde nos indica que las transacciones las cuales son detectadas como fraude son canceladas, debido a esto para nuestro análisis de detección de fraude, hay algunas columnas que no deben ser consideras, las cuales serian: **'oldbalanceOrg','newbalanceOrig','oldbalanceDest' y 'newbalanceDest'**.

* Por otro lado, consideramos que las columnas **"nameOrig"** y **"nameDest"**, al ser columnas que no tienen un patrón definido y solo representan los códigos de los clientes tanto que realizan un intercambio monetario, no serian relevantes para los análisis. Sin embargo, ya le sacamos el provecho necesario ya que creamos una nueva variable con la combinación de estas dos, la que consideramos muy importante, ya que las iniciales de los registros en estas variables nos indican si es el cliente de origen (C) o el cliente final (M).

* Por ultimo, la variable **"isFlaggedFraud"**, si bien nos indica cuando una transacción es fraudulenta, es para transacciones muy grandes de dinero, consideramos que esto ya esta incluido en la variable **"isFraud"**, y podría generar mucha correlación, la cual es más completa porque tiene todos los posibles casos. Por esta razón decidimos prescindir de esta variable.

```python
## Creamos un backup de nuestro dataframe base
df_banco_new = df_banco.clone()

## Eliminamos las variables consideradas no relevantes para el análisis
df_banco_new = df_banco_new.drop(['nameOrig','oldbalanceOrg','newbalanceOrig','nameDest','oldbalanceDest','newbalanceDest','isFlaggedFraud'])
df_banco_new
```

## <ins>Imputando valores Atípicos</ins>

* Los valores Atípicos que poseemos en la variable 'amount', pueden generarnos ciertos sesgos durante nuestra evaluación del modelo, asimismo pueden distorsionar la distribución de los datos, y darnos menos robustez en nuestros algoritmos de machine learning. Debido a todo esto, decidimos proceder con la imputación de estos valores Atípicos.
* Realizaremos la imputación con el método de Truncamiento de Percentiles.

```python
## Verificando con el método describe la presencia de valores atípicos en nuestras variables.
df_banco_new.describe().to_pandas()
```

* Tal como se habia observado en el gráfico de boxplot, la variable amount tiene presencia de valores atípicos los cuales realizaremos el tratamiento.

```python
## verificando percentiles por la parte inferior
df_banco_new.select(['amount', 'step']).describe(percentiles=[0.01,0.025,0.05,0.1,0.5,0.75]).to_pandas()
```

```python
## verificando percentiles por la parte superior
df_banco_new.select(['amount', 'step']).describe(percentiles=[0.5,0.75,0.9,0.95,0.975,0.98,0.99]).to_pandas()
```

* Observamos que hay muchos valores atípicos en la parte superior para la variable 'amount', por convención realizaremos el análisis con el $IQR = Q3 - Q1$ . De esta manera, nuestro nuevo límite de truncamiento superior será el  $P75 + 1.5*IQR$

```python
list(df_banco_new.select('amount').describe(percentiles=[0.25,0.75]).to_pandas().iloc[5:7,:]['amount'])
```

```python
## Consiguiendo el límite de truncamiento superior
p25 = list(df_banco_new.select('amount').describe(percentiles=[0.25,0.75]).to_pandas().iloc[5:7,:]['amount'])[0]
p75 = list(df_banco_new.select('amount').describe(percentiles=[0.25,0.75]).to_pandas().iloc[5:7,:]['amount'])[1]
IQR = p75 - p25
LTS = p75 + 1.5*IQR  ## Límite de Truncamiento Superior
```

```python
## Valor percentil mínimo
pmin = list(df_banco_new.select('amount').describe().to_pandas().iloc[4:5,:]['amount'])[0]
pmin
```

```python
## Imputando los valores con la función clip, acotamos entre los percentiles seleccionados
# df_banco_new['amount'] = np.clip(df_banco_new['amount'],pmin,LTS)
# Aplicar clip a la columna 'amount'
df_banco_new = df_banco_new.with_columns(pl.col("amount").apply(lambda x: min(max(x, pmin), LTS)).alias("amount"))
```

```python
## Veamos los cambios en la distribución
df_banco_new.select(['amount', 'step']).describe(percentiles=[0.5,0.75,0.9,0.95,0.975,0.98,0.99]).to_pandas()
```

# 4. CONSTRUCCIÓN DE MODELOS

Cuando utilizamos Hiperparámetros con GridSearchCV de scikit-learn, es importante definir la métrica que se utilizará para evaluar el rendimiento de los modelos al aplicar los diferentes hiperparámetros. La elección de la métrica adecuada depende de la naturaleza del problema y de los objetivos específicos que tengamos en mente. A continuación, se muestran algunas consideraciones a tener en cuenta para cada métrica al momento de utilizarlas como scoring:


> Accuracy (accuracy):

* Cuándo usarlo: Cuando todas las clases son igualmente importantes y la clase desbalanceada no es un problema significativo.
* Ejemplo: Diagnóstico de enfermedades donde todas las enfermedades son igualmente probables y tienen consecuencias similares.

> ROC AUC (roc_auc):

* Cuándo usarlo: Cuando tienes un problema de clasificación binaria y te interesa la capacidad del modelo para distinguir entre las clases positivas y negativas.
* Ejemplo: Detección de fraudes, diagnóstico de enfermedades críticas (donde es crucial minimizar falsos negativos).

> F1 Score (f1):

* Cuándo usarlo: Cuando tienes un problema de clasificación binaria con clases desbalanceadas y deseas un equilibrio entre precisión (precision) y sensibilidad (recall).
* Ejemplo: Detección de spam, donde tanto los falsos positivos como los falsos negativos tienen consecuencias significativas.

> Precision (precision):

* Cuándo usarlo: Cuando el costo de los falsos positivos es alto.
* Ejemplo: Diagnóstico médico para una enfermedad rara, donde un falso positivo puede llevar a tratamientos innecesarios.

> Recall (recall):

* Cuándo usarlo: Cuando el costo de los falsos negativos es alto.
* Ejemplo: Detección de fraude, donde un falso negativo significa que un caso de fraude pasa desapercibido.

**************************************************************************************************
**************************************************************************************************

> Validación Cruzada (Cross-Validation)

La validación cruzada divide el conjunto de datos en varias "particiones" o "folds". El proceso típico es el siguiente:

* División de los Datos: El conjunto de datos se divide en k particiones (folds) de tamaño aproximadamente igual.
* Entrenamiento y Evaluación: El modelo se entrena k veces, cada vez utilizando k-1 particiones como conjunto de entrenamiento y la partición restante como conjunto de prueba.
* Promedio de los Resultados: Se promedian los resultados de las k evaluaciones para obtener una estimación más robusta del rendimiento del modelo.

El parámetro cv en GridSearchCV define el número de particiones (folds) que se usarán en este proceso. Por ejemplo, cv=5 significa que se usarán 5 particiones.

## 4.2 Árbol de Decisión

* max_depth: Limita la profundidad máxima del árbol. Un árbol muy profundo puede captar demasiado detalle del conjunto de entrenamiento y no generalizar bien.

* min_samples_split: El número mínimo de muestras requeridas para dividir un nodo. Un valor más alto evitará divisiones en subconjuntos pequeños que podrían no ser representativos.

* min_samples_leaf: El número mínimo de muestras que debe tener un nodo hoja. Aumentar este valor puede reducir la variabilidad en el modelo.

* max_features: El número máximo de características a considerar para encontrar la mejor división. Limitar el número de características puede hacer que el modelo sea menos propenso a capturar ruido.

* ccp_alpha (Costo de Complejidad de Poda): Un parámetro para la poda de árboles, donde se eliminan las ramas que tienen un costo de complejidad proporcionalmente bajo para evitar el overfitting.

### 4.3.2 Segunda Forma

En un Random Forest, tanto max_depth como n_estimators juegan roles importantes en la configuración del modelo, pero afectan el sobreajuste y el rendimiento de diferentes maneras.

> max_depth
* Propósito: Controla la profundidad máxima de cada árbol individual en el bosque.
* Impacto en el Sobreajuste: Un max_depth muy grande puede llevar a que cada árbol se ajuste demasiado a los datos de entrenamiento, capturando mucho ruido y características específicas, lo que conduce al sobreajuste. Limitar la profundidad ayuda a reducir la complejidad de cada árbol y puede mejorar la generalización.
* Consideraciones: Ajustar max_depth a valores más bajos generalmente ayuda a reducir el sobreajuste, pero si se establece demasiado bajo, puede causar un subajuste, donde el modelo no es lo suficientemente complejo para capturar las relaciones en los datos.

> n_estimators
* Propósito: Determina el número de árboles en el bosque.
* Impacto en el Sobreajuste: Aumentar el número de árboles (n_estimators) generalmente mejora el rendimiento del modelo hasta cierto punto. Cada árbol adicional reduce la variancia del modelo, lo que puede mejorar la generalización. Sin embargo, después de un cierto número, los beneficios disminuyen y se obtiene un rendimiento decreciente.
* Consideraciones: Más árboles pueden mejorar la estabilidad del modelo y reducir el sobreajuste, pero también aumentan el tiempo de entrenamiento y el uso de memoria. Es importante encontrar un balance donde se tenga suficiente número de árboles para obtener una buena precisión sin hacer el entrenamiento prohibitivo en términos de recursos.

```python
n_estimators = [100, 200, 300]
max_depth = [None, 10, 20, 30]

for n in n_estimators:
    for m in max_depth:
        rf = RandomForestClassifier(max_depth=m, n_estimators = n).fit(x_train_scaled,y_train)
        yhat = rf.predict(x_test_scaled)
        yhat_proba = rf.predict_proba(x_test_scaled)[:,1]
        print("max_depth="+str(m)+", n_estimators="+str(n)
            +"->Accuracy: "+str(accuracy_score(y_test,yhat))
            +"->Precision: "+str(precision_score(y_test,yhat))
            +"->Recall: "+str(recall_score(y_test,yhat))
            +"->F1-score: "+str(f1_score(y_test,yhat))
            +"->AUC : "+str(roc_auc_score(y_test,yhat_proba)))
```
